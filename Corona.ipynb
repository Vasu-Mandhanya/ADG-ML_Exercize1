```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Preprocessing

```python
data = pd.read_csv('corona_dataset.csv')
data.head()
```

```python
plot=data[['Occupation','Infect_Prob']]
```

```python
data.describe()
```

```python
data.info()
```

```python
""" 
perfrom data preprocessing on 'data' Dataframe
The final dataframe must be named data
eg .... data =  {preprocessed data}    
"""

'''Name and Designation will not affect the probability of a person getting infected'''
data=data.drop(['Name','Designation'],axis=1)
```

```python
'''
Categorizing non-numerical data for regression to work
'''
data.Region=data.Region.astype("category")
data.Region=data.Region.cat.codes
data.Gender=data.Gender.astype("category")
data.Gender=data.Gender.cat.codes
data.Married=data.Married.astype("category")
data.Married=data.Married.cat.codes
data.Occupation=data.Occupation.astype("category")
data.Occupation=data.Occupation.cat.codes
data.Mode_transport=data.Mode_transport.astype("category")
data.Mode_transport=data.Mode_transport.cat.codes
data.comorbidity=data.comorbidity.astype("category")
data.comorbidity=data.comorbidity.cat.codes
data['Pulmonary score']=data['Pulmonary score'].astype("category")
data['Pulmonary score']=data['Pulmonary score'].cat.codes
data['cardiological pressure']=data['cardiological pressure'].astype("category")
data['cardiological pressure']=data['cardiological pressure'].cat.codes
```

```python
data.info()
```

```python
'''
Filling empty children column and empty FT/Month with 0 and .
'''

def zero(x):
    if np.isnan(x):
        return 0
    else:
        return x
data.Children=data.Children.apply(zero)
data['FT/month']=data['FT/month'].apply(zero)
```

```python
'''
Filling other empty cells of continuous value columns with mean
Filling Insurance increases rmse
'''
data.Diuresis=data.Diuresis.fillna(data.Diuresis.mean())
data.Platelets=data.Platelets.fillna(data.Platelets.mean())
data.HBB=data.HBB.fillna(data.HBB.mean())
data['d-dimer']=data['d-dimer'].fillna(data['d-dimer'].mean())
data['Heart rate']=data['Heart rate'].fillna(data['Heart rate'].mean())
data['HDL cholesterol']=data['HDL cholesterol'].fillna(data['HDL cholesterol'].mean())
#data.Insurance=data.Insurance.fillna(data.Insurance.mean())
```

```python
data.info()
data = data.dropna()
```

```python
""" separating features and target """

Y = data['Infect_Prob']
X = data.drop(['Infect_Prob'],axis=1)
```

```python
from sklearn.model_selection import train_test_split

train_X,test_X,train_Y,test_Y = train_test_split(X,Y)

print("Training Size : ",train_X.shape)
print("Testing Size : ",test_X.shape)
```

### Regression to predict the probability of Corona Infection

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor()
```

```python
""" regression works with numeric values,
make sure you map all the string values in the data to some numeric value if required
"""

regressor.fit(train_X,train_Y)
```

### Performance evaluation

```python
""" complete the function RMSE 
    
    The Root Mean Square Error is defined in the following way
    sqrt(summation((y_pred[i]-y_true[i])^2)/N)
    where:
         i varies from 0 to N
         
    (use numpy)     

"""
def RMSE(y_pred,y_true):   
    rmse = np.sqrt(np.mean((np.subtract(y_pred,y_true)**2)))
    return rmse
    
y_pred = regressor.predict(test_X)

print("Root Mean Square Error = ",RMSE(y_pred,test_Y))
```
